```python
 def normalize(column):
   df_mean=column.mean()
   df_std=column.std()

   return(column - df_mean)/df_std
```

```python
features=['Active','Discharged','Deaths','Active Ratio','Death Ratio']
for feature in features:
  df_copy[feature]=normalize(df_copy[feature])
```
